1. We will use a public dataset which is suitable for classification model

```python
#Importing pandas library as pd
import pandas as pd
```

```python
#Loading dataset using pd.read_csv
data = pd.read_csv("penguins_size.csv")
```

```python
#showing dataset top 5 using data.head()
data.head()
```

```python
#checking for null value
data.isnull().sum()
```

```python
data.isna().sum()
```

```python
data["culmen_length_mm"] = data["culmen_length_mm"].fillna(data["culmen_length_mm"].mean())  # Replace with mean (numerical data)
data["culmen_depth_mm"] = data["culmen_depth_mm"].fillna(data["culmen_depth_mm"].mean())  # Replace with mean (numerical data)
data["flipper_length_mm"] = data["flipper_length_mm"].fillna(data["flipper_length_mm"].mean())  # Replace with mean (numerical data)
data["body_mass_g"] = data["body_mass_g"].fillna(data["body_mass_g"].mean())  # Replace with mean (numerical data)
data["sex"] = data["sex"].fillna(data["sex"].mode()[0])  # Replace with mode
```

```python
data.isna().sum()
```

```python
#lable encoding 
from sklearn.preprocessing import LabelEncoder as le
data.species=le.fit_transform(y = data.species, self=le)
data.island=le.fit_transform(y = data.island, self=le)
data.sex=le.fit_transform(y = data.sex, self=le)
```

```python
data.head()
```

```python
#loading neccessary library for making ml models Like KNN, SVM and decision tree
from sklearn.model_selection import KFold
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
x =data[["island", "culmen_length_mm","culmen_depth_mm","flipper_length_mm","body_mass_g",	"sex"]]
y =data[["species"]]
```

```python
x.head()
```

```python
y.head()
```

```python
x_train, x_test , y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
knn_model = KNeighborsClassifier()
knn_model.fit(x_train,y_train)
```

```python
pred_knn = knn_model.predict(x_test)
```

```python
print("Accuracy:",accuracy_score(y_test,pred_knn))
print("Precision:",precision_score(y_test, pred_knn, average = 'macro'))
print("Recall:",recall_score(y_test, pred_knn, average = 'macro'))
print("F1-score:",f1_score(y_test,pred_knn,average='macro'))
```

```python
svm_model = SVC()
svm_model.fit(x_train,y_train)
```

```python
pred_svm = svm_model.predict(x_test)
```

```python
print("Accuracy:",accuracy_score(y_test,pred_svm))
print("Precision:",precision_score(y_test, pred_svm, average = 'macro'))
print("Recall:",recall_score(y_test, pred_svm, average = 'macro'))
print("F1-score:",f1_score(y_test,pred_svm,average='macro'))
```

```python
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(x_train,y_train)
```

```python
pred_decision_tree = decision_tree_model.predict(x_test)
```

```python
print("Accuracy:",accuracy_score(y_test,pred_decision_tree))
print("Precision:",precision_score(y_test, pred_decision_tree, average = 'macro'))
print("Recall:",recall_score(y_test, pred_decision_tree, average = 'macro'))
print("F1-score:",f1_score(y_test,pred_decision_tree,average='macro'))
```

```python
#for comparing with the help of average which model fits the best
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
#KNN models avg
k_average = cross_val_score(knn_model, x,y,cv=5)
k_average
```
